# EDA 

```python
# see the corelation between all the columns by using heatmap
sns.heatmap(superstore.corr(),cmap="YlGnBu",annot=True)
plt.show()
```

```python
sns.pairplot(superstore,hue="Region")
plt.show()
```

```python
superstore['Ship Mode'].value_counts()
```

```python
# Initialize a new figure
plt.figure(figsize=(8,6))

# orders placed by ship mode
sns.countplot(data=superstore,x='Ship Mode')
plt.title('check orders ship mode wise')
plt.show()
```

see maximum numbers of orders placed orders are in standard class and few numbers of oreders are in same day

```python
# average profit for each ship mode
plt.figure(figsize=(10,8))
ship1=superstore.groupby(['Ship Mode'])['Profit'].mean().reset_index()
sns.barplot(x='Ship Mode',y='Profit',data=ship1)
plt.title('average profit ship mode wise')
plt.xlabel('Ship Mode')
plt.ylabel('Profit')
plt.show()
```

orders placed by first class shipment gain the maximum profit from orders, same day and second class shipment profit are almost same and standard class shipment has less profit.

```python
# average sales for each ship mode
plt.figure(figsize=(10,8))
ship1=superstore.groupby(['Ship Mode'])['Sales'].mean().reset_index()
sns.barplot(x='Ship Mode',y='Sales',data=ship1)
plt.title('average sales ship mode wise')
plt.xlabel('Ship Mode')
plt.ylabel('Sales')
plt.show()
```

```python
superstore['Category'].value_counts()
```

```python
# orders distribution by category wise by pie chart
plt.figure(figsize=(10, 8))
plt.pie(superstore['Category'].value_counts(), labels=superstore['Category'].value_counts().index, autopct="%1.1f%%")
plt.title("Distribution of Category's")
plt.show()
```

Maximum orders almost 60% orders for office supplies, 21% for furniture and 19% for technology.  

```python
# see orders distribution for each sub-category by pie chart
plt.figure(figsize=(10, 8))
plt.pie(superstore['Sub-Category'].value_counts(), labels=superstore['Sub-Category'].value_counts().index, autopct="%1.1f%%")
plt.title("Distribution of Sub-Category")
plt.show()
```

Maximum numbers of orders sub category are binders and paper.

```python
# Create separate bar plots for Sales and Profit category-wise
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Sales bar plot
sns.barplot(x="Category", y="Sales", data=superstore, ax=axes[0])
axes[0].set_title("Sales by Category")
axes[0].set_xlabel("Category")
axes[0].set_ylabel("Sales")

# Profit bar plot
sns.barplot(x="Category", y="Profit", data=superstore, ax=axes[1])
axes[1].set_title("Profit by Category")
axes[1].set_xlabel("Category")
axes[1].set_ylabel("Profit")

# Ensure proper spacing between subplots
plt.tight_layout() 
plt.show()
```

for technology product sales and profit both are high, in case of furniture sales is higher than the office supplies category but for furniture profit is less than the office supplies category.

```python
# group all the subcategorical values 
subcategorical_sum=superstore.groupby("Sub-Category").sum()
subcategorical_sum
```

```python
# Create bar plots for Sales and Profit sub category-wise
subcategorical_sum[["Sales","Profit"]].plot(kind="bar",figsize=(20,8))
plt.title('Sales and Profit sub category wise')
```

for profuct chairs and phones total sales are very high and for copiers and phones the profit is high. In case of bookcases and tables there is negative profit.

```python
# distribution of orders for each segment
plt.figure(figsize=(10, 8))
plt.pie(superstore['Segment'].value_counts(), labels=superstore['Segment'].value_counts().index, autopct="%1.1f%%")
plt.title("Distribution the orders segment wise")
plt.show()
```

see almost 52% orders are placed by consumers, 30% are by corporate and 18% orders for home office.

```python
# Distribution of orders for each state
plt.figure(figsize=(30,20))
state_orders=sns.countplot(data=superstore,x='State')
state_orders.set_xticklabels(state_orders.get_xticklabels(),rotation=90)
plt.title('Distribution of orders state wise')
plt.show()
```

see highest number of orders are placed from the state of California followed by new york and texas.

```python
superstore.groupby("State")["Profit"].sum().sort_values(ascending=False).plot(kind='bar',figsize=[30,20])
plt.title("Distributin of sales state wise")
plt.show()
```

maximum profit earn from the state of california and new york, although good number of orders are placed from texas but the profit is negative.

```python
# set style 
sns.set(style="whitegrid")

# Initialize a new figure
fig, ax=plt.subplots(figsize=(10,6))

# create a barplot for sales 
sns.barplot(x="Region",y="Sales",data=superstore,ax=ax, color='b',label='Sales')
# create a barplot for profit
sns.barplot(x="Region",y="Profit",data=superstore,ax=ax, color='r',label='Profit')
plt.title("Sales & Profit by Region")
plt.xlabel("Region")
plt.ylabel("Values")
ax.legend()
plt.show()
```

```python
# Initialize a new figure
fig = plt.figure(figsize = (10, 4))

# Create a subplot for the Sales pie chart
plt.subplot(1, 2, 1)
superstore.groupby("Region")["Sales"].sum().plot.pie(autopct = "%1.0f%%")
plt.title("Region Vs Sales")

# Create a subplot for the Profit pie chart
plt.subplot(1, 2, 2)
superstore.groupby("Region")["Profit"].sum().plot.pie(autopct = "%1.0f%%")
plt.title("Region vs Profit")

# Add a title to the whole plot
plt.suptitle('Sales and Profit by Region')

# Display the plot
plt.show()
```

Using pie chart we also see sales are high in west and east region and profit is also high in west and east region.

```python
# Initialize a new figure
plt.figure(figsize=(8,6))

# create a bar plot for discount column region wise
sns.barplot(x="Region",y="Discount",data=superstore,palette='viridis')
plt.title("Discount by Region")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.show()
```

we previously see for west region sales and profit both are high and discount is very low. but the central region is opposite of west here sales and profit are comparetively low but the discount is high. 

```python
# Create a line plot for Profit vs. Discount
plt.figure(figsize=(8, 6))
sns.lineplot(x="Discount", y="Profit", data=superstore, marker='o', color='b')

# Add labels and a title
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.title("Profit vs. Discount")
plt.show()
```

see where we give very low discount here profit is increasing but where we give more discount there the profit is drastically reduce.

```python
# Create a line plot for Profit vs. Discount
plt.figure(figsize=(8, 6))
sns.lineplot(x="Discount", y="Sales", data=superstore, marker='o', color='b')

# Add labels and a title
plt.xlabel("Discount")
plt.ylabel("Sales")
plt.title("Sales vs. Discount")
plt.show()
```

see till 50% discount the sales are increasing but more than 50% discount the sales are very low.
